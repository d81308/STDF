# Functions

```python
## STDF Parser- created by pystdf.team
class MemoryWriter:
    def __init__(self):
        self.data = []
    def after_send(self, dataSource, data):
        self.data.append(data)
    def write(self,line):
        self.data.append(line)
    def flush(self):
        pass # Do nothing

def ImportSTDF(fname):
    with open(fname,'rb') as fin:
        p = Parser(inp=fin)
        storage = MemoryWriter()
        p.addSink(storage)
        p.parse()
    return storage.data
```

```python
# processing - created by Neil
def file_list():
    file_list=[]
    
    for file in os.listdir(os.path.abspath(os.getcwd()) + "/STDF_data"): 
    # Check whether file is in text format or not 
        if file.endswith(".stdf"): 
            file_list.append(file)
    return file_list

def data_classification(file):
    dict_classifier={}
    process_count=0
    round_count=0
    
    read_data = ImportSTDF(file_open)
    
    for datum in read_data:
        if datum[0].__class__.__name__.upper()=='PRR':
            for k,v in zip(datum[0].fieldMap,datum[1]):  #using fieldMap in pystdf to decode
                if k[0] not in dict_classifier.keys():
                    dict_classifier[k[0]]=[v]
                else:
                    dict_classifier[k[0]].append(v)
                process_count += 1
            round_count += 1
    #data_integrity_checker
    if process_count/round_count == len(dict_classifier.keys()):
        print("integrity checker: {}/{}={}".format(process_count,round_count,len(dict_classifier.keys())))
        print("data integrity is good.")
        return dict_classifier
    else:
        print("exist missing data, please check data integrity.")
    
def dict2dataframe(dict):
    df = pd.DataFrame()
    for key in dict.keys():
        df[key]=dict[key]
    return df

def viz(x,y,id):
    #data list
    x=x
    y=y
    id=id

    # plt setting
    fig = plt.figure() 
    ax = fig.add_subplot(1, 1, 1) 

    plt.title("Distribution of Points")
    plt.xlabel("X coordinate")
    plt.ylabel("Y coordinate")

    #scatter plot
    ax.scatter(x, y, color="blue", s=50, marker="o", alpha=0.2)

    #show id on grapg
    for i in range(len(x)):
        plt.annotate(id[i], xy=(x[i], y[i]), color="black", size=12, ha="center", va="center")

    #add grid    
    ax.set_yticks(np.arange(min(y)-1,max(x)+1,1))
    ax.set_yticks(np.arange(min(y)-1,max(x)+1,1)+0.5, minor=True)

    ax.set_xticks(np.arange(min(x)-1,max(x)+1,1))
    ax.set_xticks(np.arange(min(x)-1,max(x)+1,1)+0.5, minor=True)
    
    #zoom
    auto_zoom = 1
    if auto_zoom != 1:
        plt.xlim([10, 20])
        plt.ylim([-10, 0])
    
    ax.grid(True, which="minor")
    ax.set_aspect("equal")

    plt.show()
```
