```python
## import libray
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pystdf.IO import Parser
```

```python
# import file
fname="/Users/jianlin/Documents/Project_STDF_wafermap/main_Lot_1_Wafer_1_Oct_13_09h33m41s_STDF"
```

```python
# Parser function
class MemoryWriter:
    def __init__(self):
        self.data = []
    def after_send(self, dataSource, data):
        self.data.append(data)
    def write(self,line):
        self.data.append(line)
    def flush(self):
        pass # Do nothing
```

# Start to explore

```python
#read, unpack and parsing
with open(fname,'rb') as fin:
    p = Parser(inp=fin)
    storage = MemoryWriter()
    p.addSink(storage)
    p.parse()
    for line in storage.data:
        print(line)
```

```python
# unpack heaeder object - using fieldMap
with open(fname,'rb') as fin:
    p = Parser(inp=fin)
    storage = MemoryWriter()
    p.addSink(storage)
    p.parse()
    for datum in storage.data:
        print(datum[0].fieldMap,"\n")
```

```python
# filter-on: PRR section
with open(fname,'rb') as fin:
    p = Parser(inp=fin)
    storage = MemoryWriter()
    p.addSink(storage)
    p.parse()
    for datum in storage.data:
        if datum[0].__class__.__name__.upper()=='PRR':
            print(datum[0].fieldMap,"\n")
```

```python
# headers, records using zip
with open(fname,'rb') as fin:
    p = Parser(inp=fin)
    storage = MemoryWriter()
    p.addSink(storage)
    p.parse()
    for datum in storage.data:
        if datum[0].__class__.__name__.upper()=='PRR':
            for k,v in zip(datum[0].fieldMap,datum[1]):
                print(k[0],v)
```
