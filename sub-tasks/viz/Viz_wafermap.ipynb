```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [[1, 2, 'A'], [3, 4, 'B'], [5, 6, 'C']]
x = [row[0] for row in data] # 取出 X座標
y = [row[1] for row in data] # 取出 Y座標
id = [row[2] for row in data] # 取出 ID

plt.scatter(x, y) # 繪製散點圖
for i in range(len(data)):
    plt.annotate(id[i], (x[i], y[i])) # 在每個點上添加 ID
plt.grid(color='gray', linestyle='-', linewidth=0.5, alpha=0.7)
plt.show() # 顯示圖形
```
